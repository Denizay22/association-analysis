İhtiyaç duyulan kütüphanelerin importlanması

```python
import pandas as pd
import ast
```

data'yı oku

```python
data = pd.read_csv("data.csv", 
                   dtype={'period': 'string', 'stdno': 'string', 'Code': 'string', 'classname': 'string', 'classcode': 'string'})
data
```

sütun adlarını daha anlamlı bir şekilde isimlendir

```python
data.rename(columns={'stdno': 'std_no', 'Code': 'class_code', 'classname': 'class_name'}, inplace=True)
data
```

öğrenci numarası içinde harf olan kayıtları sil

```python
data = data[~data.std_no.str.contains("[a-zA-Z]").fillna(False)].reset_index().drop('index', axis=1)
data
```

```python
lenght8 = data[data.std_no.str.len() == 8]
lenght7 = data[data.std_no.str.len() == 7]
lenght7['std_no'] = lenght7.apply(lambda x: f"0{x.std_no}", axis=1)
data = pd.concat([lenght7, lenght8])
data
```

öğrenci numarasının 2, 3 ve 4. sayılarını kullanarak o öğrencinin bölüm no'sunu ayrı bir sütuna kaydet

```python
data['department_no'] = data.apply(lambda row: row.std_no[2:5], axis=1)
data.department_no = data.department_no.astype('string')
data
```

ders kodundaki gereksiz grup no bilgisini kaldır

```python
data.class_code = data.apply(lambda row: row.class_code[:7], axis=1)
data.class_code = data.class_code.astype('string')
data
```

sütunların yerini daha okunabilir olacak şekilde düzenle

```python
data = data[['std_no', 'department_no', 'class_code', 'class_name', 'period']]
data
```

bölüm kodu ve bölüm numarası eşleşmesi yapmak için yeni bir tablo oluştur. bu tablo, ana verideki öğrenci numarasından çıkartılmış olan bölüm no ve ders kodundan çıkartılmış olan bölüm kodu sütunlarını içeriyor. basit olarak groupby kullanarak her bir kod için en çok kullanılmış olan bölüm no'su değerini bularak eşleştirme yapacağız

```python
department_code_match_table = pd.DataFrame()
department_code_match_table['department_no'] = data.department_no
department_code_match_table['department_code'] = data.class_code
department_code_match_table['department_code'] = department_code_match_table.apply(lambda row: row.department_code[:3], axis=1)
department_code_match_table.department_code = department_code_match_table.department_code.astype('string')
department_code_match_table
```

elle kendi hazırladığımız okulda açık olan tüm bölüm kodunu içeren liste.(havacılık bölümü yok çünkü kayıtları incelediğimizde bölümün kodunu içeren sadece 16 kayıt olduğunu gördük)

```python
department_dict = pd.read_csv("department_dict.csv",
                           dtype={'department_code': 'string', 'faculty_name': 'string'})
department_dict
```

daha sonra filtreleme için kullanmak üzere tüm bölüm kodlarını içeren bir liste oluştur

```python
department_code_list = list(department_dict.department_code)
department_code_list
```

kayıtlardan, bölüm kodu listesinde olmayan kayıtları siliyoruz. böylece örneğin ITB, ATA gibi kodları içeren dersler tablodan kaldırılmış oluyor.

```python
department_code_match_table = department_code_match_table[department_code_match_table.department_code.isin(department_code_list)]
department_code_match_table = department_code_match_table.reset_index().drop('index', axis=1)
department_code_match_table
```

department_code üzerinde groupby yaparak, her bir department_code için en çok kullanılan department_no'yu buluyoruz. en çok kullanılan department_no'yu o bölümün bölüm kodu olarak atıyoruz

```python
department_code_match_table = department_code_match_table.groupby('department_code')['department_no'].apply(lambda x: x.value_counts().index[0]).reset_index()
department_code_match_table
```

ana tablomuzdaki department_no değerine bakıp, herhangi bir bölüme ait olmayan değer içeren satırları siliyoruz. böylece yüksek lisans tarzı kayıtlar silinmiş oluyor

```python
data = data[data.department_no.isin(list(department_code_match_table.department_no))].reset_index().drop('index', axis=1)
data
```

kendi hazırladığımız, tüm bölümlerin mesleki seçmeli derslerini içeren tablomuzu okuyoruz

```python
elective_courses = pd.read_csv("elective_courses.csv", 
                               dtype={'department_code': 'string'}, 
                               converters={"elective_courses": ast.literal_eval})
elective_courses
```

az önce elde ettiğimiz department_code_match_table'daki değerleri kullanarak elective_courses tablosuna department_no sütunu ekledik

```python
elective_courses['department_no'] = elective_courses.apply(
    lambda row: department_code_match_table[department_code_match_table.department_code == row.department_code].reset_index().at[0, 'department_no'], axis=1)
elective_courses = elective_courses[['department_code', 'department_no', 'elective_courses']]
elective_courses
```

ana tablomuza to_drop adında yeni bir sütun ekliyoruz. daha sonra bu sütundaki değeri kullanarak hangi satırları sileceğimize karar vereceğiz. her satır için to_drop değeri ise şöyle atanıyor: her satırın class_code ve department_no değeri alınır, eğer class_code değeri department_no'ya karşılık gelen mesleki seçmeli listesinde(elective_courses tablosunda) yok ise to_drop True, var ise False.

```python
data['to_drop'] = data.apply(lambda row: row.class_code not in (elective_courses[elective_courses.department_no == row.department_no].reset_index().at[0, 'elective_courses']),
                                axis=1)
data
```

to_drop sütunu True olan satırlarımızı dropluyoruz ve elimizde üzerinde analiz yapılabilecek, kullanılabilir veri kalıyor.

```python
data = data[data.to_drop == False].reset_index().drop(['index', 'to_drop'], axis=1)
data
```

```python
data.to_csv('data_cleaned.csv', index=False)
```
