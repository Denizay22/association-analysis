```python
import pandas as pd
import ast
```

Verinin Okunması

```python
data = pd.read_csv("csv/data.csv", 
                   dtype={'period': 'string', 'stdno': 'string', 'Code': 'string', 'classname': 'string', 'classcode': 'string'})
data
```

Sütun adlarının değiştirilmesi

```python
data.rename(columns={'stdno': 'std_no', 'Code': 'class_code', 'classname': 'class_name'}, inplace=True)
data
```

Öğrenci numarasında harf olan kayıtların silinmesi

```python
deleted = data[data.std_no.str.contains("[a-zA-Z]").fillna(False)].reset_index().drop('index', axis=1)
data = data[~data.std_no.str.contains("[a-zA-Z]").fillna(False)].reset_index().drop('index', axis=1)
deleted
```

```python
data
```

Öğrenci numarası uzunluğu 8 olmayan kayıtların silinmesi

```python
lenght8 = data[data.std_no.str.len() == 8]
lenght7 = data[data.std_no.str.len() == 7]
lenght7
```

```python
lenght7['std_no'] = lenght7.apply(lambda x: f"0{x.std_no}", axis=1)
data = pd.concat([lenght7, lenght8])
data
```

Öğrenci numarasındaki bölüm numarası kısmının yeni bir sütuna kaydedilmesi

```python
data['department_no'] = data.apply(lambda row: row.std_no[2:5], axis=1)
data.department_no = data.department_no.astype('string')
data = data[['std_no', 'department_no', 'class_code', 'class_name', 'period']]
data
```

Ders kodu kısmındaki gereksiz grup no kısmının silinmesi

```python
data.class_code = data.apply(lambda row: row.class_code[:7], axis=1)
data.class_code = data.class_code.astype('string')
data
```

Sütunların yerinin düzenlenmesi

```python
data = data[['std_no', 'department_no', 'class_code', 'class_name', 'period']]
data
```

İleri aşamalarda veriyi temizlemek için çeşitli dosyaların okunması

```python
department_code_match_table = pd.read_csv('csv/dp_code_match_table.csv', dtype={'department_code': 'string', 'department_no': 'string'})
```

```python
department_dict = pd.read_csv("csv/department_dict.csv",
                           dtype={'department_code': 'string', 'faculty_name': 'string'})
```

```python
department_code_list = list(department_dict.department_code)
```

Daha önce öğrenci numarasından oluşturduğumuz bölüm numarası, üniversitedeki 48 bölümden herhangi biriyle eşleşmeyen kayıtların silinmesi. Bu sayede yüksek lisans vb. tarzı kayıtlar siliniyor

```python
deleted_data = data[~data.department_no.isin(list(department_code_match_table.department_no))].reset_index().drop('index', axis=1)
data = data[data.department_no.isin(list(department_code_match_table.department_no))].reset_index().drop('index', axis=1)
deleted_data
```

```python
data
```

İlerde kullanmak için dosya okunması. Mesleki seçmeliler, Sosyal seçmeliler ve zorunlu derslerin bulunduğu liste.

```python
vocational_elective_courses = pd.read_csv("csv/vocational_elective_courses.csv", 
                               dtype={'department_code': 'string'}, 
                               converters={"vocational_elective_courses": ast.literal_eval})
vocational_elective_courses['department_no'] = vocational_elective_courses.apply(
    lambda row: department_code_match_table[department_code_match_table.department_code == row.department_code].reset_index().at[0, 'department_no'], axis=1)
vocational_elective_courses = vocational_elective_courses[['department_code', 'department_no', 'vocational_elective_courses']]
```

```python
social_elective_courses = pd.read_csv("csv/social_elective_courses.csv",
                                dtype={'department_code': 'string'},
                                converters={"social_elective_courses": ast.literal_eval})
social_elective_courses['department_no'] = social_elective_courses.apply(
    lambda row: department_code_match_table[department_code_match_table.department_code == row.department_code].reset_index().at[0, 'department_no'], axis=1)
social_elective_courses = social_elective_courses[['department_code', 'department_no', 'social_elective_courses']]
```

```python
mandatory_courses = pd.read_csv("csv/mandatory_courses.csv",
                                dtype={'department_code': 'string'},
                                converters={"mandatory_courses": ast.literal_eval})
mandatory_courses['department_no'] = mandatory_courses.apply(
    lambda row: department_code_match_table[department_code_match_table.department_code == row.department_code].reset_index().at[0, 'department_no'], axis=1)
mandatory_courses = mandatory_courses[['department_code', 'department_no', 'mandatory_courses']]
```

```python
dep_codes = vocational_elective_courses.department_code.to_list()
dep_codes.append('ITB')
```

ders kodu kısmı ITB veya bölüm kısaltmalarından herhangi biri olmayan kayıtların silinmesi

```python
data['to_drop'] = data.apply(lambda row: row.class_code[:3] not in dep_codes, axis=1)
deleted_data = data[data.to_drop == True].reset_index().drop(['index', 'to_drop'], axis=1)
data = data[data.to_drop == False].reset_index().drop(['index', 'to_drop'], axis=1)
deleted_data.class_code.unique()
```

```python
data
```

dataya öğrencinin bulunduğu bölümün kodu ve öğrencinin aldığı dersin bulunduğu bölümün kodunun eklenmesi.
daha sonra bu kodlar ile öğrenci dersi başka bir bölümden mi almış analizi yapılacak.

```python
data['department_code'] = data.apply(lambda row: 
                                    department_code_match_table[department_code_match_table.department_no == row.department_no].reset_index().at[0, 'department_code'], axis=1)
data['class_department_code'] =data.apply(lambda row: row.class_code[:3], axis=1)
data = data[['std_no', 'department_no', 'department_code', 'class_department_code', 'class_code', 'class_name', 'period']]
data.to_csv('csv/data_cleaned.csv', index=False)
data
```

tüm sosyal seçmeli, mesleki seçmeli ve zorunlu derslerin kodlarının bulunduğu bir listenin hazırlanması ve ders kodu bu listede olmayan kayıtların silinmesi. bu sayede daha önce silinmemiş olan yüksek lisans tarzı kayıtlar da tamamen silinmiş oluyor.

```python
all_classes = list()
for row in mandatory_courses.iterrows():
    all_classes += row[1][2]
for row in vocational_elective_courses.iterrows():
    all_classes += row[1][2]
for row in social_elective_courses.iterrows():
    all_classes += row[1][2]
```

```python
data['to_drop'] = data.apply(lambda row: row.class_code not in all_classes, axis=1)
deleted_data = data[data['to_drop'] == True].reset_index().drop(['index', 'to_drop'], axis=1)
data = data[data['to_drop'] == False].reset_index().drop(['index', 'to_drop'], axis=1)
deleted_data
```

```python
data
```

Mesleki seçmeli derslerin bulunduğu kayıtlar bulunarak dosyaya kaydediliyor

```python
data_voc_el = data
data_voc_el['to_drop'] = data_voc_el.apply(lambda row: row.class_code not in 
                        vocational_elective_courses[vocational_elective_courses.department_no == row.department_no].reset_index().at[0, 'vocational_elective_courses'], axis=1)
data_voc_el = data_voc_el[data_voc_el.to_drop == False].reset_index().drop(['index', 'to_drop'], axis=1)
data_voc_el
```

```python
data_voc_el.to_csv('csv/data_voc_el.csv', index=False)
```

Sosyal seçmeli derslerin bulunduğu kayıtlar bulunarak dosyaya kaydediliyor

```python
data_social_el = data
data_social_el['to_drop'] = data_social_el.apply(lambda row: row.class_code not in 
                            social_elective_courses[social_elective_courses.department_no == row.department_no].reset_index().at[0, 'social_elective_courses'], axis=1)
data_social_el = data_social_el[data_social_el.to_drop == False].reset_index().drop(['index', 'to_drop'], axis=1)
data_social_el
```

```python
data_social_el.to_csv('csv/data_social_el.csv', index=False)
```
