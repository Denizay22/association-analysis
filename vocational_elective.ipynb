```python
import pandas as pd
import numpy as np
from os import listdir
```

```python
files = listdir("meslekisecmeliler")
my_dict = {}
for file in files:
    my_dict[file[:3]] = pd.read_csv(f'csv/meslekisecmeliler\{file}', header=None)
```

```python
df = pd.DataFrame()
df['department_code'] = ""
df['vocational_elective_courses'] = ""
for k, v in my_dict.items():
    df.loc[len(df.index)] = [k, v[0].to_list()]
df
```

```python
df.to_csv('csv/vocational_elective_courses.csv', index=False)
```
